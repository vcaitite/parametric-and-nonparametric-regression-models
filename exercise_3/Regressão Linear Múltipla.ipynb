## Atividade 1:

Mostre que ao assumir um modelo: $Y_i \sim Normal(\mu_{(x_i)} = \beta_0 + \beta_1 * x_i; \sigma^2)$, sendo a priori dada por: $\beta_1 \sim Normal(\mu_1; \sigma_1 ^2)$. A posteriori para o parâmetro $\beta_1$ segue a forma: 

$\beta_1 \sim Normal(\frac{\tau \sum_i (y_i - \beta_0)x_i + \tau_1 \mu_1}{\tau(\sum_i x_i^2) + \tau_1};  \frac{1}{\tau(\sum_i x_i^2) + \tau_1})$, 

sendo $\tau = 1/\sigma^2$ e $\tau_1 = 1/\sigma_1^2$.

### Resolução:

A posteriori é expressa como: $P(H|D) = \frac{[P(D|H) * P(H)]}{P(D)}$, onde P(H|D) representa a probabilidade a posteriori, P(D|H) é a verossimilhança, P(H) é a probabilidade a priori e P(D) é a probabilidade total dos dados.

A verossimilhança conjunta para $Y = (Y_1, Y_2, \dots, Y_n)$ dado $\beta_0, \beta_1$ e $\sigma^2$ é:

$p(Y \mid \beta_0, \beta_1) = \prod_{i=1}^n \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{(Y_i - \beta_0 - \beta_1 x_i)^2}{2 \sigma^2}\right).$

Como estamos interessados apenas na posteriori de $\beta_1$, podemos omitir os termos constantes e escrever a verossimilhança em termos de $\beta_1$ apenas:

$p(Y \mid \beta_1) \propto \exp\left(-\frac{1}{2 \sigma^2} \sum_{i=1}^n (Y_i - \beta_0 - \beta_1 x_i)^2\right)$.

Expandindo:

- $p(Y \mid \beta_1) \propto \exp\left(-\frac{1}{2 \sigma^2} \sum_{i=1}^n \left[(Y_i - \beta_0)^2 - 2 (Y_i - \beta_0) \beta_1 x_i + \beta_1^2 x_i^2\right]\right).$

- $p(Y \mid \beta_1) \propto \exp\left(-\frac{1}{2 \sigma^2} \left(\sum_{i=1}^n (Y_i - \beta_0)^2 - 2 \beta_1 \sum_{i=1}^n (Y_i - \beta_0) x_i + \beta_1^2 \sum_{i=1}^n x_i^2\right)\right)$.

Considerando apenas os termos que envolvem $\beta_1$:

- $p(Y \mid \beta_1) \propto \exp\left(-\frac{\beta_1^2}{2 \sigma^2} \sum_{i=1}^n x_i^2 + \frac{\beta_1}{\sigma^2} \sum_{i=1}^n (Y_i - \beta_0) x_i\right).$

Sabe-se que a priori para $\beta_1$ é:

$\beta_1 \sim \text{Normal}(\mu_1, \sigma_1^2)$

Logo:

$p(\beta_1) \propto \exp\left(-\frac{1}{2 \sigma_1^2} (\beta_1 - \mu_1)^2\right).$

Reescrevendo:

$p(\beta_1) \propto \exp\left(-\frac{\tau_1}{2} (\beta_1 - \mu_1)^2\right).$

Fazendo o produto da verossimilhança com a probabilidade a priori, temos:

- $p(\beta_1 \mid Y) \propto p(Y \mid \beta_1) \cdot p(\beta_1).$

- $p(\beta_1 \mid Y) \propto \exp\left(-\frac{\beta_1^2}{2 \sigma^2} \sum_{i=1}^n x_i^2 + \frac{\beta_1}{\sigma^2} \sum_{i=1}^n (Y_i - \beta_0) x_i\right) \cdot \exp\left(-\frac{\tau_1}{2} (\beta_1 - \mu_1)^2\right).$

- $p(\beta_1 \mid Y) \propto \exp\left(-\frac{\beta_1^2}{2 \sigma^2} \sum_{i=1}^n x_i^2 + \frac{\beta_1}{\sigma^2} \sum_{i=1}^n (Y_i - \beta_0) x_i - \frac{\tau_1}{2} (\beta_1^2 - 2 \beta_1 \mu_1 + \mu_1^2)\right).$

- $p(\beta_1 \mid Y) \propto \exp\left(-\frac{1}{2} \left(\beta_1^2 \tau \left(\sum_{i=1}^n x_i^2 + \tau_1\right) - 2 \beta_1 \tau \left(\sum_{i=1}^n (Y_i - \beta_0) x_i + \tau_1 \mu_1\right)\right)\right).$

Para comparar com a densidade de uma normal, observamos que a expressão dentro da exponencial é quadrática em $\beta_1$. Reescrevendo, temos:

$-\frac{1}{2} \left( A \beta_1^2 - 2 B \beta_1 \right),$

onde

$A = \tau \sum_{i=1}^n x_i^2 + \tau_1$

e

$B = \tau \sum_{i=1}^n (Y_i - \beta_0) x_i + \tau_1 \mu_1$.

Essa forma quadrática é característica de uma distribuição normal. A densidade de uma normal ($\text{Normal}(\text{média}, \text{variância})$) pode ser escrita na forma:

$\exp\left(-\frac{1}{2 \sigma^2} (\beta_1 - \mu)^2\right)$, onde $\mu$ é a média e $\sigma^2$ é a variância. 

Assim, para identificar a média e a variância da posteriori de $\beta_1$, basta comparar os coeficientes da expressão quadrática em $\beta_1$ que obtivemos.

Reecrevendo: 

- $A \beta_1^2 - 2 B \beta_1 = A \left(\beta_1^2 - 2 \frac{B}{A} \beta_1\right) = A \left[\left(\beta_1 - \frac{B}{A}\right)^2 - \left(\frac{B}{A}\right)^2\right]$.

Substituindo de volta na expressão original, temos:

- $p(\beta_1 \mid Y) \propto \exp\left(-\frac{A}{2} \left(\left(\beta_1 - \frac{B}{A}\right)^2 - \left(\frac{B}{A}\right)^2\right)\right).$

Simplificando:

- $p(\beta_1 \mid Y) \propto \exp\left(-\frac{A}{2} \left(\beta_1 - \frac{B}{A}\right)^2\right).$

Assim, a média e a variância são dadas por:

- A **média** da posteriori para $\beta_1$ é $\frac{B}{A}$, ou seja:
  
  $\mu_{\beta_1} = \frac{\tau \sum_{i=1}^n (Y_i - \beta_0) x_i + \tau_1 \mu_1}{\tau \sum_{i=1}^n x_i^2 + \tau_1}$.
  

- A **variância** da posteriori para $\beta_1$ é $\frac{1}{A}$, ou seja:

  $\sigma_{\beta_1}^2 = \frac{1}{\tau \sum_{i=1}^n x_i^2 + \tau_1}$.

Logo a posteriori para $\beta_1$ pode ser escrita como:

$\beta_1 \mid Y \sim \text{Normal}\left(\frac{\tau \sum_{i=1}^n (Y_i - \beta_0) x_i + \tau_1 \mu_1}{\tau \sum_{i=1}^n x_i^2 + \tau_1}, \frac{1}{\tau \sum_{i=1}^n x_i^2 + \tau_1}\right).$

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
from math import *
import statsmodels.api as sm
```

```python
dataset = pd.read_csv('~/Documents/UFMG/Doutorado/materias/Modelos_de_Regressao/atividade_3/DEA_VariaveisAmbientais.csv', sep=",", engine='python')
dataset.head(5)
```

```python
dataset.columns.values
```

```python
# Selecionando as variáveis de ambiente requeridas:
df = dataset[["PrecipMedAnualMed","DensFocQueim",
                "DeclMedPerc","AltMedVeg",
                "FocQueim","DescAtmMed",
                "TempMedAn","UnidCons_TI",
                "PercEstraPavQA","PercEstraPavIN"]]

y = dataset["DEA.Media21"]
```

### Modelo de Regressão Linear considerando todas as variáveis de ambiente

```python
df = sm.add_constant(df)
model = sm.OLS(y, df)
results = model.fit() 
print(results.summary())
```

### Remoção de Variáveis do Modelo com P-valor > 0.05


Inicialmente todas as variáveis com P-valor > 0.05 serão eliminadas do modelo.

```python
df.drop(["DensFocQueim", "DeclMedPerc", "AltMedVeg", "DescAtmMed", "TempMedAn", "UnidCons_TI"], axis=1, inplace=True)
df.head()
```

As variáveis de ambiente selecionadas foram:
    
- PrecipMedAnualMed (Precipitation)
- FocQueim (Fire spots)
- PercEstraPavQA (Road access points)
- PercEstraPavIN (Road intersections)

```python
model = sm.OLS(y, df)
results = model.fit() 
print(results.summary())
```

### Remoção de Variáveis do Modelo que obtiveram correlação diferente da esperada


Removendo também a variável "Road Access Points" do modelo pois ela está apresentando uma correlação negativa (diferentemente do correlação esperada apresentada na tabela do exercício).

As variáveis de ambiente selecionadas foram:

### Remoção da Variável do Modelo com P-valor > 0.05

Remoção da variável PercEstraPavIN (Road intersections) pois possui P-valor > 0.05.

```python
df.drop(["PercEstraPavIN"], axis=1, inplace=True)
df.head()
```

```python
model = sm.OLS(y, df)
results = model.fit() 
print(results.summary())
```

### Plot do Modelo Final

```python
x, y = model.exog_names[1:]

x_range = np.arange(df[x].min(), df[x].max(), 10)
y_range = np.arange(df[y].min(), df[y].max(), 10)

X, Y = np.meshgrid(x_range, y_range)

exog = pd.DataFrame({x: X.ravel(), y: Y.ravel()})
Z = results.predict(exog = sm.add_constant(exog)).values.reshape(X.shape)
```

```python
fig = plt.figure(figsize=plt.figaspect(1)*2)
ax = plt.axes(projection='3d')
z = dataset["DEA.Media21"]
ax.scatter(df[x].values, df[y].values, z.values, label="Data")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, alpha = 0.4)
ax.legend()
ax.view_init(elev=20, azim=30) #Reproduce view
plt.show()
```
